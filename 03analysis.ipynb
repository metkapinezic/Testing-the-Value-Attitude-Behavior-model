```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assuming df is your DataFrame
correlation_matrix = df.corr()

# Create a matrix to store p-values
p_values = np.zeros_like(correlation_matrix)

# Calculate correlation coefficients and p-values
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr, p_value = pearsonr(df[correlation_matrix.columns[i]], df[correlation_matrix.columns[j]])
        correlation_matrix.iloc[i, j] = corr
        p_values[i, j] = p_value

# Create a DataFrame for p-values
p_values_df = pd.DataFrame(p_values, index=correlation_matrix.columns, columns=correlation_matrix.columns)

# Plot the correlation matrix with p-values
fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)

# Add the p-values as text
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        if i != j:
            plt.text(i, j, f"{correlation_matrix.iloc[i, j]:.2f}\n({p_values_df.iloc[i, j]:.3f})",
                     ha='center', va='center', color='black', fontsize=10)

# Set axis labels
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title('Correlation Matrix with Statistical Significance')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Assuming df is your DataFrame
selected_variables = ['AVG-V-Self-Enhacemen', 'AVG-V-Openness_to_change', 'AVG-V-Self_transcendence',
                       'AVG-V-Conservation', 'AVG-P-Extraversion', 'AVG-P-Honesty-Humility', 'AVG-ATT',
                       'BEH-take_part', 'BEH-passenger-usage', 'BEH-driver-usage', 'INT-passenger', 'INT-driver']

# Select only the specified variables from the DataFrame
df_selected = df[selected_variables]

# Calculate the correlation matrix for the selected variables
correlation_matrix = df_selected.corr()

# Create a matrix to store p-values
p_values = np.zeros_like(correlation_matrix)

# Calculate correlation coefficients and p-values
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        corr, p_value = pearsonr(df_selected[correlation_matrix.columns[i]], df_selected[correlation_matrix.columns[j]])
        correlation_matrix.iloc[i, j] = corr
        p_values[i, j] = p_value

# Create a mask for significant correlations
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)  # Upper triangle mask

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(10, 8))

# Draw the heatmap with significant correlations
cax = sns.heatmap(correlation_matrix, mask=mask, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt=".2f", cbar_kws={"shrink": 0.8})

# Add significance labels
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if p_values[i, j] < 0.05:  # Adjust significance level as needed
            significance = "*"
        elif p_values[i, j] < 0.01:
            significance = "**"
        elif p_values[i, j] < 0.001:
            significance = "***"
        else:
            significance = ""  # No significance label if p-value is not significant

        ax.text(j + 0.5, i + 0.5, significance, ha='center', va='center', color='black', fontsize=12)

# Set axis labels
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title('Correlation Matrix for Selected Variables with Significance Labels')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from tabulate import tabulate

# Assuming df is your DataFrame
correlation_matrix = df.corr()

# Create a matrix to store p-values
p_values = np.zeros_like(correlation_matrix)

# Calculate correlation coefficients and p-values
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr, p_value = pearsonr(df[correlation_matrix.columns[i]], df[correlation_matrix.columns[j]])
        correlation_matrix.iloc[i, j] = corr
        p_values[i, j] = p_value

# Create a DataFrame for p-values
p_values_df = pd.DataFrame(p_values, index=correlation_matrix.columns, columns=correlation_matrix.columns)

# Create a table with correlation coefficients and p-values
table_data = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        correlation_coefficient = correlation_matrix.iloc[i, j]
        p_value = p_values_df.iloc[i, j]
        significance = ""
        
        if p_value < 0.05:
            significance = "*"
        if p_value < 0.01:
            significance = "**"
        if p_value < 0.001:
            significance = "***"

        table_data.append([correlation_matrix.columns[i], correlation_matrix.columns[j],
                           correlation_coefficient, p_value, significance])

# Print the table
headers = ["Variable 1", "Variable 2", "Correlation Coefficient", "P-Value", "Significance"]
table = tabulate(table_data, headers=headers, floatfmt=(".2f", ".2f", ".3f", ".3f", ""), tablefmt="pretty")
print(table)
```
