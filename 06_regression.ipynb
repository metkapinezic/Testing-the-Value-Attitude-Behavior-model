Do Values influence attitude

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_csv('cleandata.csv')

# Define independent variables (features)
independent_vars = [
    'AVG-V-Self-Enhacemen', 'AVG-V-Openness_to_change', 'AVG-V-Self_transcendence',
    'AVG-V-Conservation', 'AVG-P-Opennes_to_experience', 'AVG-P-Extraversion',
    'AVG-P-Honesty-Humility'
]

# Define the dependent variable (target)
dependent_var = 'AVG-ATT'

# Extract independent and dependent variables from the dataset
X = data[independent_vars]
y = data[dependent_var]

# Add a constant to the independent variables (for intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression model
print(model.summary())

```

These results are from a multiple linear regression analysis, where the dependent variable is 'AVG-ATT' and the independent variables are 'AVG-V-Self-Enhacemen', 'AVG-V-Openness_to_change', 'AVG-V-Self_transcendence', 'AVG-V-Conservation', 'AVG-P-Opennes_to_experience', 'AVG-P-Extraversion', and 'AVG-P-Honesty-Humility'.

Here's how to interpret the key parts of the output:

R-squared: This is a measure of how well the independent variables explain the variation in the dependent variable. In this case, the R-squared value is 0.093, which means that approximately 9.3% of the variance in 'AVG-ATT' is explained by the independent variables.

Adj. R-squared: This is the R-squared adjusted for the number of predictors in the model. It penalizes the addition of unnecessary predictors to the model. The adjusted R-squared is 0.055.

F-statistic: This is a measure of overall significance of the model. It tests whether at least one of the independent variables is significantly related to the dependent variable. Here, the F-statistic is 2.483 with a corresponding p-value of 0.0188, suggesting that the model is statistically significant at the 0.05 level.

Coefficients: These represent the estimated regression coefficients for each independent variable. Each coefficient indicates the change in the dependent variable for a one-unit change in the corresponding independent variable, holding all other variables constant. For example, the coefficient for 'AVG-V-Self_transcendence' is 0.2087, which means that for each one-unit increase in 'AVG-V-Self_transcendence', 'AVG-ATT' is expected to increase by 0.2087 units.

t-statistic and p-value: These indicate the statistical significance of each coefficient. A low p-value (typically below 0.05) suggests that the coefficient is statistically significant. In this output, 'AVG-V-Self_transcendence' has a statistically significant coefficient with a p-value of 0.002.

Omnibus, Durbin-Watson, Jarque-Bera, Skewness, Kurtosis: These are tests for the normality of the residuals and other diagnostic statistics. They help to assess whether the assumptions of linear regression are met.

Notes: This section provides additional information about the model, including the assumption about the covariance matrix of the errors.

Overall, the model suggests that 'AVG-V-Self_transcendence' is the most significant predictor of 'AVG-ATT' among the variables included in the model. However, the overall explanatory power of the model (R-squared) is relatively low, indicating that there may be other factors not accounted for in the model.

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_csv('cleandata.csv')

# Define independent variables (features)
independent_vars = [
    'AVG-V-Self-Enhacemen', 'AVG-V-Openness_to_change', 'AVG-V-Self_transcendence',
    'AVG-V-Conservation', 'AVG-P-Opennes_to_experience', 'AVG-P-Extraversion',
    'AVG-P-Honesty-Humility'
]

# Define the dependent variable (target)
dependent_var = 'AVG-ATT'

# Iterate through each independent variable and fit separate regression models
for var in independent_vars:
    # Extract independent and dependent variables from the dataset for the current variable
    X = data[var]
    y = data[dependent_var]

    # Add a constant to the independent variable (for intercept)
    X = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(y, X).fit()

    # Print the summary of the regression model for the current variable
    print(f"Regression results for {var}:")
    print(model.summary())
    print("\n")
```
